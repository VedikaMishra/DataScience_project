```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("credit_customers (DS).csv")
df.head()
```

```python
#Data Preprocessing
df.isnull().sum()
```

```python
df.info()
```

```python
#Handling duplicate records
df.duplicated()
```

```python
#Their Is No Duplicacy in above Dataframe
```

```python
# removing the unnecessary columns
df=df.drop(['personal_status','other_parties','other_payment_plans','own_telephone','foreign_worker','purpose','checking_status'],axis=1)
```

```python
df.head()
```

```python
print(df.shape)
```

```python
#select x and y
x=df.iloc[:,:-1] # every column except the last i.e. class column
y=df.iloc[:,-1] # only class column
print(type(x))
print(type(y))
print(x.shape)
print(y.shape)
df.head()
```

```python
#splitting the data into train_test
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_train.shape)
df.head()

```

```python
from sklearn.preprocessing import LabelEncoder

# for other attributes
le_credit_history= LabelEncoder()
le_savings_status=LabelEncoder()
le_employment=LabelEncoder()
le_property_magnitude=LabelEncoder()
le_housing=LabelEncoder()
le_job = LabelEncoder()

le_class = LabelEncoder() # for target attribute

# Renaming the converted attributes

df['credit_history_n']= le_credit_history.fit_transform(df['credit_history'])
df['savings_status_n'] = le_savings_status.fit_transform(df['savings_status'])
df['employment_n'] = le_employment.fit_transform(df['employment'])
df['property_magnitude_n'] = le_property_magnitude.fit_transform(df['property_magnitude'])
df['housing_n'] = le_housing.fit_transform(df['housing'])
df['job_n'] = le_job.fit_transform(df['job'])

df['class_n'] = le_class.fit_transform(df['class'])

df.drop(['credit_history','savings_status','employment','property_magnitude','housing','job','class'], axis= 'columns' , inplace=True)
df.head()
```

```python
x = df.iloc[:,:-1] #dep
y = df.iloc[:,-1]  #ind
print(type(x))
print(type(y))
print(x.shape)
print(y.shape)
```

```python
#splitting the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#logisticRegression
from sklearn.linear_model import LogisticRegression
m1= LogisticRegression(solver="liblinear")
m1.fit(x_train,y_train)
#accuracy
print('Train_score',m1.score(x_train,y_train))
print('Test_score',m1.score(x_test,y_test))
```

```python
# predicting class for test dataset
ypred_m1 = m1.predict(x_test)
print(ypred_m1)
```

```python
# generating confusion matrix & classification report
from sklearn.metrics import confusion_matrix, classification_report
confusion_m1 = confusion_matrix(y_test,ypred_m1)
print(confusion_m1)
print(classification_report(y_test,ypred_m1))
#calculating the accuracy
accuracy_model1 = m1.score(x_test,y_test)
accuracy_model1
```

```python
# Applying KNN Classifer
from sklearn.neighbors import KNeighborsClassifier

m2 = KNeighborsClassifier(n_neighbors=85)
m2.fit(x_train,y_train)
# calculating the accuracy
print('Train_score',m2.score(x_train,y_train))
print('Test_score',m2.score(x_test,y_test))
```

```python
# predicting class for test dataset
ypred_m2 = m2.predict(x_test)
print(ypred_m2)
```

```python
# generating confusion matrix and classification report

confusion_m2 = confusion_matrix(y_test,ypred_m2)
print(confusion_m2)
print(classification_report(y_test,ypred_m2))
#calculating the accuracy
accuracy_model2 = m2.score(x_test,y_test)
accuracy_model2
```

```python
# SVM with linear(SVC) kernel
from sklearn.svm import SVC
m3_linear = SVC(kernel = 'linear',C= 10)
m3_rbf = SVC(kernel = 'rbf', C=1.0, gamma='scale')
m3_linear.fit(x_train,y_train)
m3_rbf.fit(x_train,y_train)
#accuracy with linear kernel
print('Train_score',m3_linear.score(x_train,y_train))
print('Test_score',m3_linear.score(x_test,y_test))
#accuracy with rbf kernel
print('Train_score',m3_rbf.score(x_train,y_train))
print('Test_score',m3_rbf.score(x_test,y_test))
```

```python
# predicting class for test dataset
ypred_m3_linear = m3_linear.predict(x_test) # with linear
print(ypred_m3_linear)
ypred_m3_rbf = m3_rbf.predict(x_test) #rbf
print(ypred_m3_rbf)
```

```python
# generating confusion matrix and classification report
# with linear kernel
confusion_m3_lin = confusion_matrix(y_test,ypred_m3_linear)
print(confusion_m3_lin)
print(classification_report(y_test,ypred_m3_linear))
#calculating the accuracy
accuracy_model3_lin = m3_linear.score(x_test,y_test)
accuracy_model3_lin
```

```python
# with linear kernel
confusion_m3_rbf = confusion_matrix(y_test,ypred_m3_rbf)
print(confusion_m3_rbf)
print(classification_report(y_test,ypred_m3_rbf))
#calculating the accuracy
accuracy_model3_rbf = m3_rbf.score(x_test,y_test)
accuracy_model3_rbf
```

```python
# Reporting the best acurracy
best_model = max(accuracy_model1,accuracy_model2,accuracy_model3_lin,accuracy_model3_rbf)
if (best_model == accuracy_model1):
    print("Logistic Regression has the best accuracy of", accuracy_model1 * 100 ," % for the given dataset")
if (best_model == accuracy_model2):
    print("KNN has the best accuracy of", accuracy_model2 * 100," % for the given dataset")
if (best_model == accuracy_model3_lin):
    print("SVM has the best accuracy of", accuracy_model3_lin * 100,"% with linear kernel for the given dataset")
if (best_model == accuracy_model3_rbf):
    print("SVM has the best accuracy of", accuracy_model3_rbf * 100,"% with RBF kernel for the given dataset")
```
